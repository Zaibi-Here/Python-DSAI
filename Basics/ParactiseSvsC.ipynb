```python
#Even or Odd Numbers
Num = 59
if Num % 2 == 0:    # If the number divided by 2 has no remainder, it's Even
    print("Even")
else:
    print("Odd")
```

```python
# Largest of 3 Numbers
Num1 = int(input("Enter 1st Number: "))
Num2 = int(input("Enter 2nd Number: "))
Num3 = int(input("Enter 3rd Number: "))

#Check the LARGEST
if Num1 >= Num2 and Num1 >= Num3:
    print("Largest:", Num1)
elif Num2 >= Num1 and Num2 >= Num3:
    print("Largest:", Num2)
else:
    print("Largest:", Num3)
```

```python
# Find the Palindrome
xt = "12321"
# print(xt)
print(xt[2])
if xt[0]==xt[4] and xt[1]==xt[3]:
    print ("Palindrome")
else:
    print ("Not a Palindrome")


# 2nd Method Find the Palindrome
xt = "12321"
Rxt = xt[::-1]
print(Rxt)

if xt == Rxt:
    print("Yes Palindrome")
else:
    print("No")
```

```python
# Find the Vowels From the String
ww = "This is the Simple String to Findout The Vowels AEIOU"
count = 0

vowels = "aeiouAEIOU"

for i in ww:
    if i in vowels: #Loops to Checks Vowels
        count += 1

print(count)



```
