```python
# # X = Dependent Variable
# # Y = Independent Variable
import pandas as pd

# Example dataset creation
data = {
    'BMXWT': [32, 34, 54, 45, 76, 67, 47, 38, 19, 99],
    'BMXHT': [4, 5, 7, 6, 9, 12, 14, 10, 3, 2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define X and y
X = df[['BMXWT']]  # Independent Variable (Weights)
y = df['BMXHT']  # Dependent Variable (Heights)
```

**Splitting Dataset into Training+Testing Data**

```python
X.head()
```

```python
#Import Sckit learn Library
# from sklearn.linear_model import linearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
# test_size mean Division of Data
```

```python
from sklearn.linear_model import LinearRegression #LinearRegression From SkLearn

model = LinearRegression() #create the Model
model.fit(X_train, y_train) #Fit with the Model
```

```python
#Plotting Train Data
import matplotlib.pyplot as plt
# plt.scatter(X_test, y_test, color='red')
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train), color='Green')
plt.title('Height vs Weight (Training set)')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
#Plotting Test Data
import matplotlib.pyplot as plt
# plt.scatter(X_test, y_test, color='red')
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, model.predict(X_test), color='Green')
plt.title('Height vs Weight (Training set)')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
#Testing my Model - Model Fitness - Test
model.score(X_test, y_test)
```

```python
#Testing my Model - Model Fitness - Train
model.score(X_test, y_test)
```
