# **Using Most Common Values**

```python
import numpy as np   #Import Numpy and Find Means

sizes = np.array([23,23,12,32,43,54,65,43,34,76,89,43,56,78,54,32])
sizes.mean()
```

```python
from collections import Counter
cnt = Counter()

for size in sizes:     #to find Counter of each Value
  cnt[size] += 1

cnt.most_common()  #Most Common Values
```

```python
from scipy import stats
stats.mode(sizes) #The Most Occour Values in all of it
```
