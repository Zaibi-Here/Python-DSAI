```python
#Works on Functions like np.sum(), np.mean(), np.min(), np.max(), np.std(), etc.
array_hai_bhai = np.array ([2,3,4,5,6,4,8,5,21])

print(array_hai_bhai + 2) #adding 2 to Each Number
print(array_hai_bhai * 5) #Multiplying 2 to Each Number

print(np.sum(array_hai_bhai))   #to find SUM
print(np.mean(array_hai_bhai))  #to find MEAN
print(np.max(array_hai_bhai))  #to find MAX
print(np.min(array_hai_bhai)) #to find MIN
print(np.std(array_hai_bhai)) #to find STD
print(np.sqrt(array_hai_bhai))  #to find SQRT
```

```python
#Multiply and Inverse the Matrix
Matrix1 = np.array([[5,7],[7,9]])
Matrix2 = np.array([[4,6],[5,9]])

Product = np.matmul(Matrix1,Matrix2) #Multiply the 2 Matrics
print(Product)

Inverse = np.linalg.inv(Matrix2) #Inverse any 1 Matrix
print(Inverse)
```

```python
#Merge Concatenate Data Frames
my1 = pd.DataFrame ({"X" : [5,8],"Y" : [3,6]})
my2 = pd.DataFrame ({"Z" : [0,4],"W" : [4,9]})

result = pd.concat([my1,my2])
result
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],     #Merges the Two Matrics
                     'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
merged = pd.merge(left, right, on='key')
print(merged)
```

```python
#Make a Groups By Checking
data = {
    "Name": ["Ali","Anjum","Shoaib","Raja"],
    "Age": [23,21,25,15],
    "Grade": ['A','F','D','C'],
    "Kills": [455,324,675,454],
    "Date": ['2024-09-20', '2024-09-21', '2024-09-22','2030-09-20']

}

df = pd.DataFrame(data)

resulx = df.groupby('Age')['Kills'].mean()
resulx
```

```python
mysort = df.sort_values(by="Kills", ascending=False)
mysort

# SSort the Kills in a Decending Order
```
