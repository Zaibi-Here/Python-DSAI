```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.linspace(0, 10, 100)   #Adding the Data
y = np.sin(x)
z = np.cos(x)

#3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D line
ax.plot(x, y, z, color='Orange')

# Labels
ax.set_title('Simple 3D Line Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)  # Grid X,Y
z = np.sin(np.sqrt(x**2 + y**2))  # Func X,Y

# 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
#cmap= plasma,inferno,coolwarm
# Labels
ax.set_title('Complex 3D Surface Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# right side color bar
fig.colorbar(surf)

plt.show()
```
